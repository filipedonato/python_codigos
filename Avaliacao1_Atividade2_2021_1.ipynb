```python
#Importar as bibliotecas
from control.matlab import *
from numpy import *
import matplotlib.pyplot as plt

#Passar as matrizes que estão nos cálculos para o código
A = matrix([[-2 , 1 ],
           [0  , -3]])
B = matrix([[1],
            [1]])
c = matrix([[1,0]])
d = matrix([[1]])
Hss = ss(A,B,c,d)
print(Hss)              #Imprime as matrizes pra ver que escrevi correto

H = ss2tf(Hss)          #Encontrar a função de transferência
print(H)                #Mostra a função de transferência

#Agora vamos encontrar os polos e zeros
print('\n')
print('Zeros do sistema:')
zeros = zero(H)
print(zeros)

print('\n')
print('Polos do sistema:')
polos = pole(H)
print(polos)

print('\n')
print('Os autovalores da matriz A são:')
autovaloresEautovetores = linalg.eig(A)
print(autovaloresEautovetores[0])

print('\n')
zeros , polos = pzmap(H)
plt.title('Polos  zeros do sistema')
#Fim da letra B

```

```python
#Escrever a entrada que é o degrau unitário
y,t,x = step(Hss, return_x=True)      #Hss é das variáveis de estado

#Plotar agora a resposta ao degrau
plt.figure(1)
plt.plot(t,y) 
plt.xlabel('Tempo, t') 
plt.ylabel('Resposta ao degrau, y')
plt.title('Figura 1')
plt.grid(True)
plt.show()


plt.figure(2)
plt.plot(t,x) 
plt.xlabel('Tempo, t') 
plt.ylabel('Resposta ao degrau, x')
plt.title('Figura 2 - Estados X1 e X2')
plt.legend(['x1','x2'])
plt.grid(True)
plt.show()

#Fim do código
```
