```python
#Resposta Letra C
from control.matlab import *
import numpy as np
import matplotlib.pyplot as plt
import cmath
import math
from cmath import polar, phase, sqrt, pi, tan, exp
from math import degrees
from numpy import conj

cm = 1/2.54  # centimetros em polegadas

s = tf('s')

#Escreve a função G sem o K.
G = 1
H = (s+9) / (s**2 + 2*s + 3.25)
#S0 é o ponto desejado
s0 = -15 + 1j
H_s0 = (s0+9) / (s0**2 + 2*s0 + 3.25)
print('\n O argumento Arg(H_s0) = ', math.degrees(cmath.phase(H_s0)),'graus')

ArgGs0 = -degrees(phase(H(s0))) + 180.0
print('\n A fase de G(s0) é:', ArgGs0, 'graus') 
#Se for menor que zero é PI e maior que zero será PID

#Valores que calculei no papel para o projeto
ki = 7.222
kd = 0.828
kp = 45.6

#Função G(s) do controlador PID
Gs = (kd*s**2 + kp*s + ki) / s
print('\n A função de tranferência G(s) é :', Gs)

L = H*Gs
#print('\n A Função de transferencia L(s) é:', L)
#print('\n Os polos de L(s) são:', pole(L),'\n\n')
#print('\n Os zeros de L(s) são:', zero(L),'\n\n')

#Função de Transferência de Malha Fechada
FTMF = feedback(L, 1, sign=-1) #realimentação negativa
print('\n Função de tranferência de malha fechada FTMF:', FTMF)
print('\n Zeros do sistema:', zero(FTMF),)
print('\n Polos do sistema:', pole(FTMF), '\n')

plt.figure(3,figsize=(15*cm, 15*cm))
rlocus(L)
plt.title('RLOCUS')
plt.legend([L])
plt.show()

plt.figure(4)
p, z = pzmap(FTMF)
plt.title('Diagrama de Pólos e Zeros')
plt.xlabel('Real, x') 
plt.ylabel('Imaginário, j')
plt.grid(True)
plt.show()

y,t = step(FTMF, T=5)
plt.figure(5)
plt.plot(t,y) 
plt.title('Resposta ao degrau')
plt.xlabel('Tempo, t') 
plt.ylabel('Resposta ao degrau em malha fechada')
plt.grid(True)
plt.show()

print('\nCONCLUSÃO, nos pólos deu uma pequena diferença devido os arredondamentos')
```
